```python
import numpy as np
```

```python
n1=100
n2=1000
n3=10000
```

```python
x1=np.random.normal(size=n1)
x2=np.random.normal(size=n2)
x3=np.random.normal(size=n3)
```

```python
beta=[2,3]
```

```python
from scipy import stats
import pandas as pd
# Step 1, empty lists to save estimate:
est1 = []
est2 = []
est3 = []
# Step 2: for-loop:
for i in range(1000):
  # Step 3: simulate output:
    y1=beta[0]+beta[1]*x1+np.random.normal(size=n1) 
    y2=beta[0]+beta[1]*x2+np.random.normal(size=n2)
    y3=beta[0]+beta[1]*x3+np.random.normal(size=n3)
    
    slope1, intercept1, r1, p1, std_err1 = stats.linregress(x1, y1)
    slope2, intercept2, r2, p2, std_err2 = stats.linregress(x2, y2)
    slope3, intercept3, r3, p3, std_err3 = stats.linregress(x3, y3)
    

  # Step 4: append `slope`s to estimates
    est1.append(slope1)
    est2.append(slope2)
    est3.append(slope3)

# Step 5: create the DataFrame (outside of the for-loop)
esti1 = pd.DataFrame(est1)
esti2 = pd.DataFrame(est2)
esti3 = pd.DataFrame(est3)
```

```python
#Draw density of estimators for three n's
import seaborn as sns
from matplotlib import pyplot as plt
sns.set_style('whitegrid')
fig, ax = plt.subplots()
sns.kdeplot(data=esti1.squeeze(),ax=ax, color='red',label='n=100')
sns.kdeplot(esti2.squeeze(),ax=ax, color='blue',label='n=1000')
sns.kdeplot(esti3.squeeze(),ax=ax, color='black',label='n=10000')
ax.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
